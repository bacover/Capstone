- Installment Credit: predetermined length and end date such as mortgage, car loan
- Revolving Credit: No payment plan  higher interest rates such as credit card loan

- Amortizing loan:Amortization is the process of spreading out a loan into a series of fixed payments. The loan is paid off at the end of the payment schedule

- loan default: A loan goes into default—which is the eventual consequence of extended payment delinquency
- loan delinquent: Payment delinquency is commonly used to describe a situation in which a borrower misses a single payment

- loan settlement is the act of paying back the amount of money owed to the lender
- loan collection is the process of pursuing payments of debts owed by individuals or businesses

- 'Fully Paid', 'Current', 'Late', 'Grace period', 'Default', 'Charged Off'
- late to default  before charged off
- default: a note goes into Default status when it is 121 or more days past due.
- Charge-Off:When a note is in Default status, Charge Off occurs no later than 150 days past due (i.e. No later than 30 days after the Default status is reached) when there is no reasonable expectation of sufficient payment to prevent the charge off.
