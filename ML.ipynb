```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 90)
pd.set_option('display.max_rows',90)
```

```python
#df = pd.read_csv('clean_file.csv')
df = pd.read_csv("revised_accepted_2013_to_2018Q4.csv")#.iloc[:, 2:]
print(df.shape)
df.head()
#df.drop(columns='member_id',inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
randomForest= RandomForestClassifier()
```

```python
randomForest.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python


X=df.drop(columns=['loan_status2'])  # Features
#X = pd.get_dummies(df.drop(columns=['loan_status2']))
#y=pd.get_dummies(df['loan_status2']  )# Labels
y=df['loan_status2']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7,stratify=y)
```

```python
N = int(np.sqrt(X_train.shape[1]) )
N
```

```python
r = [*list(range(1, N+1,2)),X_train.shape[1]]
r
```

```python
grid_para_forest = [{
    'n_estimators':[25,50],
    "min_samples_leaf": range(1, 10,2),
    "min_samples_split": np.linspace(start=2, stop=30, num=10, dtype=int),
    'random_state':[42],
    'max_features':r,
    
}]
grid_search_forest = GridSearchCV(randomForest,grid_para_forest,scoring='accuracy',cv=5)
%time grid_search_forest.fit(X_train, y_train)
```
